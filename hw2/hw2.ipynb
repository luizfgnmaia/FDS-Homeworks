```python
import pandas as pd
import numpy as np
from dfply import * # https://github.com/kieferk/dfply
import matplotlib.pyplot as plt
import scipy # https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.pearsonr.html
from sklearn.linear_model import LinearRegression # https://towardsdatascience.com/linear-regression-in-6-lines-of-python-5e1d0cd05b8d
from sklearn.linear_model import LogisticRegression # https://www.quora.com/Does-scikit-learn-support-ordinal-logistic-regression
from sklearn.model_selection import cross_val_score # https://scikit-learn.org/stable/modules/cross_validation.html
import statistics
```

# Parte 1: pr√©-processamento dos dados

## Lendo e agrupando os dados

### ENEM2015.csv

```python
ENEM2015 = pd.read_csv('./Data/ENEM2015.csv')
ENEM2015.columns = ['COD', 'N_PART', 'Y']
ENEM2015
```

### DadosEscolares1996-2015.csv

Selecionando apenas o ano mais recente para cada escola.

```python
DadosEscolares = pd.read_csv('./Data/DadosEscolares1996-2015.csv') >> rename(COD = X.CODMEC) >> arrange(X.ANO, ascending = False) >> distinct(X.COD)
data = ENEM2015 >> inner_join(DadosEscolares, by = 'COD')
data
```

Infelizmente perdemos boa parte das entradas com esse join.

Vamos selecionar as colunas que parecem ser boas preditoras (a partir do dicion√°rio e do .csv).

```python
data = data >> select(X.COD, X.Y, X.TIPOESC, X.ZONA, X.N_PART)
data
```

### ESC2013_RMSP_CEM.csv

```python
ESC2013 = pd.read_csv('./Data/ESC2013_RMSP_CEM.csv')
data = ESC2013 >> rename(COD = X.CODESC) >> select(X.COD, X.ENEM2013, X.IDEB13AF, X.INSE_AB) >> right_join(data, by = 'COD')
data
```

```python
data.isnull().sum()
```

N√£o faz muito sentido manter a vari√°vel do IDEB dos anos finais devido ao grande n√∫mero de NaN, manterei as outras vari√°veis por enquanto.

```python
data = data.drop(columns = 'IDEB13AF')
data
```

### RendimentoEscolar2000-2015.csv

```python
RendimentoEscolar = pd.read_csv('./Data/RendimentoEscolar2000-2015.csv') >> rename(COD = X.CODMEC) >> distinct(X.COD) >> select(X.COD, X.TOTAL_EM_REPROV, X.TOTAL_EM_APROV)
data = data >> left_join(RendimentoEscolar, by = 'COD')
data.isnull().sum()
```

Tamb√©m temos bastantes NaNs nessas colunas, optarei por n√£o us√°-las.

```python
data = data.drop(columns = ['TOTAL_EM_REPROV', 'TOTAL_EM_APROV'])
data
```

## Vari√°veis categ√≥ricas

Agrupando algumas categorias e criando dummies.

```python
data['Y'].value_counts()
```

Mesmo tendo apenas uma observa√ß√£o com classifica√ß√£o 0, optei por deix√°-la para n√£o alterar a realidade da base de dados.

```python
data.Y.replace(0, 1, inplace = True)
```

```python
data['TIPOESC'].value_counts()
```

```python
data['ZONA'].value_counts()
```

```python
data.TIPOESC.replace(['EE', 'EMEFM', 'ETF'], ['OUTROS', 'OUTROS', 'OUTROS'], inplace = True)
```

```python
data['TIPOESC'].value_counts()
```

```python
data.TIPOESC.replace(['ESC.PART.', 'OUTROS'], [1, 0], inplace = True)
data = data >> mutate(SUL = X.ZONA, NORTE = X.ZONA, OESTE = X.ZONA, CENTRO = X.ZONA)
data = data.drop(columns = 'ZONA')
data.SUL.replace(['SUL', 'NORTE', 'LESTE', 'OESTE', 'CENTRO'], [1, 0, 0, 0, 0], inplace = True)
data.NORTE.replace(['NORTE', 'SUL', 'LESTE', 'OESTE', 'CENTRO'], [1, 0, 0, 0, 0], inplace = True)
data.OESTE.replace(['OESTE', 'SUL', 'NORTE', 'LESTE', 'CENTRO'], [1, 0, 0, 0, 0], inplace = True)
data.CENTRO.replace(['CENTRO', 'SUL', 'NORTE', 'LESTE', 'OESTE'], [1, 0, 0, 0, 0], inplace = True)
data
```

## Vari√°veis num√©ricas

### Gr√°ficos de dispers√£o da vari√°vel resposta contra as vari√°veis num√©ricas

```python
data.plot(kind = 'scatter', x = 'N_PART', y = 'Y')
np.corrcoef(data.Y.values, data.N_PART.values)
scipy.stats.pearsonr(data.Y, data.N_PART)
```

```python
data.plot(kind = 'scatter', x = 'ENEM2013', y = 'Y')
tmp = data >> select(X.Y, X.ENEM2013)
tmp = tmp.dropna()
np.corrcoef(tmp.Y.values, tmp.ENEM2013.values)
```

```python
data.plot(kind = 'scatter', x = 'INSE_AB', y = 'Y')
tmp = data >> select(X.Y, X.INSE_AB)
tmp = tmp.dropna()
np.corrcoef(tmp.Y.values, tmp.INSE_AB.values)
```

Podemos ver que tanto ENEM2013 quanto INSE_AB t√™m uma forte correla√ß√£o com a vari√°vel resposta enquanto N_PART tem uma correla√ß√£o n√£o t√£o alta mas ainda significativa.

### Imputa√ß√£o de valores faltantes

Como os valores faltantes est√£o nas vari√°veis cont√≠nuas ENEM2013 e INSE_AB, utilizarei um modelo de regress√£o linear com as outras vari√°veis com o objetivo de prever e inputar os valores faltantes.

```python
data_inse = data.drop(columns = ['ENEM2013', 'COD']).dropna()
Y = data_inse.INSE_AB
X = data_inse.drop(columns = ['Y','INSE_AB'])

lm_inse = LinearRegression()
lm_inse.fit(X, Y)

for i in range(len(data)):
    if pd.isna(data.INSE_AB[i]):
        pred = pd.DataFrame(columns = X.columns)
        pred.loc[0] = [data.TIPOESC[i], data.N_PART[i], data.SUL[i], data.NORTE[i], data.OESTE[i], data.CENTRO[i]]
        data.INSE_AB[i] = lm_inse.predict(pred)
```

```python
data_enem = data.drop(columns = ['COD']).dropna()
Y = data_enem.ENEM2013
X = data_enem.drop(columns = ['Y', 'ENEM2013'])

lm_enem = LinearRegression()
lm_enem.fit(X, Y)

for i in range(len(data)):
    if pd.isna(data.ENEM2013[i]):
        pred = pd.DataFrame(columns = X.columns)
        pred.loc[0] = [data.INSE_AB[i], data.TIPOESC[i], data.N_PART[i], data.SUL[i], data.NORTE[i], data.OESTE[i], data.CENTRO[i]]
        data.ENEM2013[i] = lm_enem.predict(pred)
```

# Parte 2: Modelagem

Avaliarei o ajuste de um Modelo Multinomial Logit atrav√©s da 5-Fold Cross-Validation, este modelo foi escolhido pela natureza da vari√°vel resposta e da quantidade de observa√ß√µes dispon√≠veis.

```python
Y = data.Y
X = data.drop(columns = ['COD', 'Y'])
model = LogisticRegression(multi_class = 'multinomial', solver = 'newton-cg', max_iter = 10000, fit_intercept = True)
scores = cross_val_score(model, X, Y, cv = 5)
scores
```

A m√©dia da acur√°cia nos folds √©:

```python
statistics.mean(scores)
```
